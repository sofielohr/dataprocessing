## Verantwoording statistieken

**Bronvermelding:** alle cijfers, indien anders aangegeven, zijn afkomstig van het CBS. 

**Sentiment** cijfers zijn een saldo van het gewogen percentage ondernemers met een positieve verwachting minus het gewogen percentage ondernemers met een negatieve verwachting. Dit saldo geeft inzicht in de te verwachten ontwikkeling in de komende drie maanden.

**Vacatures** is gedefinieerd als het totaal aantal vacatures waar nog geen personeel voor is gevonden. Hieronder vallen ook vacatures voor uitzendkrachten of ander tijdelijk personeel.

**Oprichtingen** is gedefinieerd als het totaal aantal oprichtingen waarbij voldaan is aan de economische kriteria voor een bedrijf. Oprichtingen betreft het onstaan van een nieuw bedrijf, niet de voortzetting van één of meerdere bedrijven. 

**Faillisementen** is gedefineerd als het aantal uitgesproken faillisementen per periode. Dit gaat om het totale aantal natuurlijke personen met eenmanszaak en bedrijven en instellingen dat door uitspraak van de rechter in staat van faillissement is verklaard.

**Omzet** wordt gemeten aan de hand van de omzetindex (2015=100). De omzet index heeft betrekking op alle bedrijven in de populatie (alle werknemersklassen).

**Bedrijven** omvat alle bedrijven met meer dan 2 werkzame personen. 

**Eenmanszaken** omvat alle bedrijven met 1 of 2 werkzame personen. 

**Ziekteverzuim** is het totaal aantal ziektedagen van de werknemers, in procenten van het totaal aantal beschikbare (werk-/kalender)dagen van de werknemers in de verslagperiode. Het percentage heeft betrekking op de volledige populatie van bedrijven (alle werknemersklasse).

**Werknemers** is in duizendtallen en heeft betrekking op de volledige populatie van bedrijven (alle werknemersklasse). Werknemers wordt onderverdeeld in vaste werknemers en flexwerkers. Onder vaste werknemers vallen alle werknemers met een contract voor onbepaalde tijd met een vast aantal uren per week. Onder flexwerkers vallen onderandere werknemers met tijdelijke contracten, uitzend-, oproep- of invalkrachten. 

**Loonsom** is in miljoenen en heeft betrekking op de volledige populatie van bedrijven (alle werknemersklassen). 

**Loonsom per fte** (full time employee) is in duizendtallen en is gedefinieerd als de gemiddelde loonsom per voltijd werknemer. De loonsom per fte wordt berekend door totale loonsom te delen door het aantal arbeidsjaren, ook wel voltijdequivalenten (vte) genoemd. 

**Zelfstandigen** is in duizendtallen. Zelfstandigen wordt opgedeeld naar zelfstandige met personeel (zmp) en zelfstandige zonder personeel (zzp).

**AOV-dekkingsgraad** is gedefinieerd als het percentage zelfstandigen met een AOV-verzekering. Dit is gebaseerd op cijfers van het CBS en analyse van Baken Adviesgroep. Het percentage heeft betrekking op de volledige populatie van bedrijven (alle werknemersklasse).

**Zelfstandigen, voornaamste inkomen** is gedefinieerd als alle personen waarbij het inkomen uit werkzaamheden als zelfstandige de voornaamste inkomensbron is.

**Inkomen zelfstandigen, voornaamste inkomen** is gedefinieerd als de mediaan van alle inkomens van de zelfstandigen, voornaamste inkomen. Het inkomen van een enkele zelfstandige is gedefinieerd als het inkomen uit eigen onderneming, loon directeuren en overige inkomsten uit arbeid. 

**Consumentenvertrouwen** is gedefineerd als het gemiddelde van de saldi van de percentages van positieve en negatieve antwoorden op de vragen over de economische situatie in de afgelopen en komende 12 maanden, de financiële situatie van het huishouden in de afgelopen en komende 12 maanden en of het een gunstige tijd is om grote aankopen te doen. De indicator kan een waarde aannemen van -100 (iedereen antwoordt negatief) tot +100 (iedereen antwoordt positief). Bij een waarde van 0 is het aandeel pessimisten gelijk aan het aandeel optimisten. Dit is de (seizoens)gecorrigeerde reeks.

**Wagenpark particulier** is gedefinieerd als wagens op naam van een natuurlijk persoon.

**Wagenpark zakelijk** is gedefinieerd als wagens op naam van een rechtspersoon (bedrijf of instelling).

**Wagenpark bedrijfsauto's** is gedefinieerd als de som van personenauto's zakelijk, bestelauto's en vrachtauto's.
